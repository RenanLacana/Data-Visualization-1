```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.grid(True)
eixo.plot(df['data'], df['temperatura'], color='g')
eixo.set_title('Temperatura no Tempo', fontsize=25)
eixo.set_xlabel('Data', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.legend(['Temperatura'], loc="lower right", fontsize=15)

x2 = df['data'][df['temperatura'].idxmax()]
y2 = df['temperatura'].max()
x1 = df['data'][df['temperatura'].idxmax() - 7000]
y1 = y2 - 5

eixo.annotate("Máximo", fontsize = 20,
            xy=(x2, y2), xytext=(x1, y1),
            arrowprops=dict(facecolor = 'k')
            )

x2 = df['data'][df['temperatura'].idxmin()]
y2 = df['temperatura'].min()
x1 = df['data'][df['temperatura'].idxmin() - 7000]
y1 = y2 + 5

eixo.annotate("Mínimo", fontsize = 20,
            xy=(x2, y2), xytext=(x1, y1),
            arrowprops=dict(facecolor = 'k')
            )


eixo.axhline(max(df['temperatura']), color='b', linestyle='--')
eixo.axhline(min(df['temperatura']), color='b', linestyle='--')
```

```python
temperatura_media_por_dia_da_semana = df.groupby('dia_da_semana')['temperatura']mean()
```

```python
temperatura_media_por_dia_da_semana
```

```python
dias = ['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']
```

```python
temperatura_media_por_dia_da_semana = temperatura_media_por_dia_da_semana[dias]
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])
indice = range(len(temperatura_media_por_dia_da_semana))
cores = ['midnightblue', 'mediumblue', 'darkslateblue', 'rebeccapurple', 'darkorchid', 'thistle', 'purple']
ax.bar(indice,temperatura_media_por_dia_da_semana, color=cores, edgecolor='k')
ax.set_title('Temperatura por dia da semana', fontsize=15,pad=10)
ax.set_ylabel('Temperatura', fontsize=15)
ax.set_xlabel('Dia da Semana', fontsize=15)
ax.set_xticks(indice)
ax.set_xticklabels(dias)
plt.show()
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])
explodir = [0.1,0,0,0,0,0,0.1]
ax.pie(temperatura_media_por_dia_da_semana, labels=temperatura_media_por_dia_da_semana.index, autopct='%.1f%%', explode=explodir, shadow=True)
ax.set_title('Temperatura por Dia da Semana', fontsize=20, pad=10)

plt.show()
```
